```python
from neuralprophet import NeuralProphet
```

```python
import pandas as pd
#from neuralprophet import NeuralProphet
from matplotlib import pyplot as plt
import pickle
```

```python
df = pd.read_csv('d:/Mubassira/AirPassengers.csv')
df.head()
```

```python
len(df)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,14)
plt.plot(df['Month'], df['#Passengers'])
plt.show()
```

```python
data = df[['Month', '#Passengers']] 
data.dropna(inplace=True)
data.columns = ['ds', 'y'] 
data.head()
```

```python
m = NeuralProphet()
```

```python
df=data
```

```python
m = NeuralProphet()
df_train, df_val = m.split_df(df, freq='M', valid_p = 0.2)
metrics = m.fit(df_train, freq='M', validation_df=df_val, progress='bar')
```

```python
metrics
```

```python
future = m.make_future_dataframe(df, periods=24, n_historic_predictions=len(df))
forecast = m.predict(future)
forecast
```

```python
fig_forecast = m.plot(forecast)
```

```python
## Visualize component
```

```python
fig_components = m.plot_components(forecast)
```
